<b>Problem statement:</b>
Find sentiments of news headlines and then tell stock will go up or down.
The goal of this project is to develop a machine learning model that can analyze news headlines to determine their sentiment and predict the impact on stock prices. The sentiment analysis will categorize news headlines into positive, negative, or neutral sentiments and assess their correlation with stock price movements.

```python
import pandas as pd
```

```python
df=pd.read_csv(r'Data.csv',encoding='ISO-8859-1')
```

```python
df.head() #0 and 1 in label refers to stock will decrese and increase repectively
```

```python
df.shape
```

```python
#training and Testing
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']  #old in training  and new data in Test
```

```python
train.shape,test.shape
```

```python
#remove punctuation using regex
data=train.iloc[:,2:27]   #2 column to 27th column
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
data
```

```python
#rename columns for ease
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=list1
```

```python
data.head(5)
```

```python
#convert into lower case 
for i in range(len(data.columns)):
    data[i]=data[i].str.lower()
data.head(1)
```

```python
#combine all lower case words of 1st row  into a sentence for the sake of vectors

' '.join(str(x)for x in data.iloc[1,0:25]) #1st row and all 25 columns
```

```python
#combine words of all rows into vectors

headlines=[]
for row in range(0,len(data.index)):   #iterating to all rows 
    headlines.append(' '.join (str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[1]
```

```python
#count vectorizer or we can use TF-IDF to convert sentences into vectors
from sklearn.feature_extraction.text import CountVectorizer 

from sklearn.ensemble import RandomForestClassifier  #can use naieve bias or multi-naive bias also
```

```python
count_vec=CountVectorizer(ngram_range=(2,2))   #2-2 words vector
train_dataset=count_vec.fit_transform(headlines)
```

```python
train_dataset[0]  #sparse matrix means 0 and 1 so it'll be vector
```

```python
#implement random forest
random_forest=RandomForestClassifier(n_estimators=200,criterion='entropy')
random_forest.fit(train_dataset,train['Label'])  #tarin['label '] is our target valriable 0 or 1
```

```python
train_dataset[0]
```

```python
#predict for the Test data
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))

test_dataset=count_vec.transform(test_transform)
predictions=random_forest.predict(test_dataset)
```

```python
predictions
```

```python
test.loc[3723,:]
```

```python
#Check accuracy and metrics

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
```

```python
score=accuracy_score(test['Label'],predictions)
print(score)
```

```python
report=classification_report(test['Label'],predictions)
print(report)
```
